```python
def draw_91_95_AWU_Avial_ratio(prefix,index):
    df = pd.read_csv('../../applications/%s/data/rnn_input_data_AWU_Avial_ratio.csv' % prefix)
    # df['ItemShortName'] == 'TC7SZ08FU'

    df = df.drop(['ItemShortName'], axis=1)
    columns = df.columns
    df
    
    AWU_Avial_ratio = df.iloc[index,:].values
    AWU_Avial_ratio
    
    seqgan_df = pd.read_csv('../../applications/%s/SeqGAN/seqgan_simliar_rnn_output.csv' % prefix)
    seqgan_df = seqgan_df.drop(['ItemShortName','clustered_label'], axis=1)
    latest5_df = seqgan_df.loc[index,['2017-03-02','2017-03-03','2017-03-06','2017-03-10','2017-03-13']]
    latest5_df = latest5_df.rename({'2017-03-02': 92, '2017-03-03': 93,'2017-03-06': 94,'2017-03-10': 95, '2017-03-13': 96 })
    # latest5_df.columns = [92,93,94,95,96]
    print(latest5_df)

    x_list = []
    y_list = []
    std_list = []
    print(type(latest5_df))
    for i in latest5_df.index:
        x_list.append(i)
        if latest5_df[i] == 0:
            y_list.append(0)
            std_list.append(0)
        else:
            data = draw(prefix, latest5_df[i])
            y_list.append(data['AWU_Avial_ratio_mean'].values[0])
            std_list.append(data['AWU_Avial_ratio_std'].values[0])
    #     print(latest5_df[i])

    print(x_list)
    print(y_list)
    print(std_list)
    
    
    import matplotlib.pyplot as plt
    plt.figure(num=3, figsize=(16, 16),)

    plt.plot(range(0,len(columns)), AWU_Avial_ratio, c ='gray',label='actual value')
    plt.scatter(range(0,len(columns)), AWU_Avial_ratio, c ='gray',label='actual value')


    plt.scatter(x_list, y_list, c ='red', marker='o',label='simulate value',s=20)
    plt.scatter(x_list, np.sum([y_list,std_list], axis = 0) , c ='blue',marker='+',label='simulate value')
    print(np.sum([y_list,std_list]))
    start = len(columns)-5
    end = len(columns)
    for i in range(start,end):
        print(i+1)
        plt.text(i+1, AWU_Avial_ratio[i], str(i+1), fontsize=10)
    #     plt.text(actual_AWU_Avial_ratio[i]+0.018, simulate_df.iloc[i]['Backlog_Avail_ratio'], str(text[i]), fontsize=10)
    # plt.xlim(0,1)
    plt.ylim(-0.01,1)
    plt.show()
```
