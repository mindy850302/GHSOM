```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

def draw(prefix, cluster):
    
    df = pd.read_csv('../../applications/%s/data/rnn_input_item_seq_with_cluster_integer.csv' % prefix)
    # df
    predict_cluster = cluster

    df_cluster = pd.read_csv('../../applications/%s/data/%s_with_clustered_label_integer_actual_value.csv' % (prefix,prefix))
    cluster = df_cluster[df_cluster['clustered_label'] == predict_cluster]
#     print(cluster)
    simulate_value = []
    columns = ['Actual_AWU','BL<=9WKs','Backlog','DCOH','Hub_OH','FCST_M']
    size = 0
    if cluster.shape[0] >= 15:
        size = 15
    else:
        size = cluster.shape[0]
    for column in columns:
        # 常態分佈抽樣 利用平均數與標準差進行抽樣

        mu = np.mean(cluster[column])
        sigma = np.std(cluster[column])
        value = []
        while True:
            value = np.random.normal(mu, sigma, size)
            if value.all() >= 0:
                break

#         print('%s: %s' % (column,value[0]))
        simulate_value.append(value)

#     print('simulate_value:',np.array(simulate_value).transpose())
    simulate_df = pd.DataFrame(data =np.array(simulate_value).transpose(), columns = columns)
#     simulate_df
    
    simulate_df['TTL_OH'] = simulate_df['DCOH'] + simulate_df['Hub_OH']
    simulate_df['Avail'] = simulate_df['TTL_OH'] + simulate_df['Backlog']
    simulate_df['AWU_Avial_ratio'] = simulate_df['Actual_AWU'] / simulate_df['Avail']
    simulate_df['Backlog_Avail_ratio'] = simulate_df['Backlog'] / simulate_df['Avail']
#     simulate_df
    
#     # cluster AWU_Avial_ratio mean
#     cluster_aar_mean = np.mean(cluster['AWU_Avial_ratio'])
#     cluster_aar_std = np.std(cluster['AWU_Avial_ratio'])
#     print('AWU_Avial_ratio mean: %s' % cluster_aar_mean)
#     print('AWU_Avial_ratio std: %s' % cluster_aar_std)

#     # cluster Backlog_Avail_ratio mean
#     cluster_bar_mean = np.mean(cluster['AWU_Avial_ratio'])
#     cluster_bar_std = np.std(cluster['Backlog_Avail_ratio'])
#     print('Backlog_Avail_ratio mean: %s' % cluster_bar_mean)
#     print('Backlog_Avail_ratio std: %s' % cluster_bar_std)
    
#     actual_AWU_Avial_ratio = cluster.sample(n=1)['AWU_Avial_ratio']
#     actual_Backlog_Avail_ratio = cluster.sample(n=1)['Backlog_Avail_ratio']
    
    actual_AWU_Avial_ratio = []
    actual_Backlog_Avail_ratio = []
    text =[]
    for i in range(0,size):   
        actual_AWU_Avial_ratio.append(cluster.iloc[i]['AWU_Avial_ratio'])
        actual_Backlog_Avail_ratio.append(cluster.iloc[i]['Backlog_Avail_ratio'])
        text.append(i+1)
#     print(actual_AWU_Avial_ratio)
#     print(actual_Backlog_Avail_ratio)
    
    cluster_aar_x = [cluster_aar_mean+cluster_aar_std,cluster_aar_mean-cluster_aar_std]
    cluster_aar_y = [cluster_bar_mean,cluster_bar_mean]
    cluster_bar_x = [cluster_aar_mean,cluster_aar_mean]
    cluster_bar_y = [cluster_bar_mean+cluster_bar_std,cluster_bar_mean-cluster_bar_std]

    cluster_x = cluster_aar_mean
    cluster_y = cluster_bar_mean
    
    
    plt.figure(num=3, figsize=(12, 12),)

#     plt.plot(cluster_aar_x, cluster_aar_y,linewidth=3, label='AWU_Avial_ratio')
#     plt.plot(cluster_bar_x, cluster_bar_y,linewidth=3, label='Backlog_Avail_ratio')

    plt.scatter(actual_AWU_Avial_ratio, actual_Backlog_Avail_ratio, c ='red',label='actual value')
    for i in range(0,len(text)):
        plt.text(actual_AWU_Avial_ratio[i], actual_Backlog_Avail_ratio[i], str(text[i]), fontsize=10)
        plt.text(simulate_df.iloc[i]['AWU_Avial_ratio']+0.018, simulate_df.iloc[i]['Backlog_Avail_ratio'], str(text[i]), fontsize=10)
    plt.scatter(simulate_df['AWU_Avial_ratio']+0.018, simulate_df['Backlog_Avail_ratio'], c ='green',label='simulate value')
    
#     plt.scatter(cluster_x, cluster_y, c ='purple',label='%s' % predict_cluster)

    x = np.arange(0,1,0.1)
    print(x)
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.xlabel('AWU_Avial_ratio')
    plt.ylabel('Backlog_Avail_ratio')
    plt.legend(loc='upper right')
    # plt.xlabel()
    # plt.legend(plots, ('Apple', 'Facebook', 'Google'), loc='best', framealpha=0.5, prop={'size': 'large', 'family': 'monospace'})
    plt.show()
```
