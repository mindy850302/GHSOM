```python
import pandas as pd 
prefix = 'wm5-normalize-experiment-6'
# prefix = 'wm5_minutes_normalize'
df = pd.read_csv('../../applications/%s/data/rnn_input_item_seq_with_cluster_integer.csv' % prefix)
len(df.clustered_label.unique())
```

```python
df_grade = pd.read_csv('../../raw-data/wm5-grade-clean.csv')
```

```python
df_grade = df_grade[['id','original','semester']]
df_grade
```

```python
merge = pd.merge(df,df_grade, on='id')
merge.to_csv('../../applications/%s/data/rnn_input_item_seq_with_cluster_integer_with_grade.csv'% prefix)
merge
```

```python
merge['semester_level'] = merge.semester//10
clusters = merge.clustered_label.unique()
clusters
```

```python
merge
```

```python
import numpy as np
grade_p_list = []
for i in range(4):
    clusters = merge.clustered_label.unique()
    for cluster in clusters:
    #     grade_count = [0] * merge[merge.clustered_label == cluster].semester_level.unique()
    #     print(merge[merge.clustered_label == cluster].semester_level.unique())
        print(cluster)
        grade_p = [0] * 10 
        print(merge[merge.clustered_label == cluster].semester_level.unique())
        for grade in merge[merge.clustered_label == cluster].semester_level.unique():
            grade_p[grade] = sum(merge[merge.clustered_label == cluster].semester_level == grade)/len(merge[merge.clustered_label == cluster].semester_level == grade)
        
        grade_p.append(cluster*pow(10,i))
        grade_p.append(len(merge[merge.clustered_label == cluster]))
        grade_p.append(np.mean(merge[merge.clustered_label == cluster].semester))
        grade_p.append(np.std(merge[merge.clustered_label == cluster].semester))
        print(grade_p)
        grade_p_list.append(grade_p)
        
    merge.clustered_label = merge.clustered_label//10
```

```python
grade_p_list
```

```python
cluster_level = pd.DataFrame(data=grade_p_list,columns=['0_9','10_19','20_29','30_39','40_49','50_59','60_69','70_79','80_89','90_99','cluster','cluster_count','mean','std'])
cluster_level
```

```python
cluster_level.drop_duplicates().sort_values(by=['cluster']).reset_index(drop=True)
```

```python
cluster_level.drop_duplicates().sort_values(by=['cluster']).reset_index(drop=True).to_csv('../../applications/%s/data/cluster_level.csv'% prefix)
```
