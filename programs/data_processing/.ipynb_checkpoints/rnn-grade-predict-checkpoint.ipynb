```python
# df = pd.read_csv('../../applications/wm5_reshape_less/data/rnn_input_item_seq_with_cluster_integer.csv')
# df = pd.read_csv('../../raw-data/wm5-raw.csv')
df = pd.read_csv('../../applications/wm5_minutes_normalize/data/rnn_input_item_seq_with_cluster_integer.csv')

wm5_grade = pd.read_csv('../../raw-data/wm5-grade.csv')
wm5_id = pd.read_csv('../../raw-data/id.csv')

df = pd.merge( df, wm5_id, on='id', how='right')

df =df.join(wm5_grade.set_index('id'), on='id')
df
```

```python
# grade = 'original'
grade = 'semester'

criteria = [df[grade].between(0, 59.9), df[grade].between(60, 69.9), df[grade].between(70, 79.9),df[grade].between(80, 89.9),df[grade].between(90, 100)]
values = [4,3,2,1,0]
df['class'] = np.select(criteria, values, 0)
df
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()

sc=MinMaxScaler()

target = 'class'
from_column = 'week15'
to_column = 'week18'
####################### 
# for normalization
####################### 
# x = (df.drop(['id',target,'Group','Group Project','Exam1','Exam2','Final','Attendance','TA','semester','original','Teacher_video_week18','Teacher_text_week18','TA_video_week18','TA_text_week18'], axis=1)/60).values
# y = df[target].values.reshape(-1, 1)

# x=sc.fit_transform(x)
# # y=sc.fit_transform(y)


# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True)


####################### 
# no normalization
####################### 
train, test = train_test_split(df, test_size=0.2, shuffle=True)

x_train_raw = train.copy()
x_test_raw = test.copy()

x_train  = train.loc[:,from_column:to_column]
x_test  = test.loc[:,from_column:to_column]

y_train =train[target]
y_test =test[target]
```

```python
x_train = (np.arange(x_train.as_matrix().max()) == x_train.as_matrix()[...,None]-1).astype(int)
y_train = (np.arange(y_train.as_matrix().max()) == y_train.as_matrix()[...,None]-1).astype(int)

# x_test = x_test.as_matrix().reshape((len(x_test), 1, x_test.shape[1]))
x_test = (np.arange(train.loc[:,from_column:to_column].as_matrix().max()) == x_test.as_matrix()[...,None]-1)
y_test = (np.arange(test.as_matrix().max()) == y_test.as_matrix()[...,None]-1).astype(int)
#y_test.as_matrix().reshape((len(y_test), 1, 1))

print("X_train:",x_train.shape)
print("Y_train:",y_train.shape)
print("X_val:",x_test.shape)
print("Y_val:",y_test.shape)
```

```python
# x_train = x_train.reshape((len(x_train), 1, x_train.shape[1]))
# y_train = y_train.reshape((len(y_train), 1, 1))

# x_test = x_test.reshape((len(x_test), 1, x_test.shape[1]))
# y_test = y_test.reshape((len(y_test), 1, 1))


x_train = x_train.as_matrix().reshape((len(x_train), 1, x_train.shape[1]))
y_train = y_train.as_matrix().reshape((len(y_train), 1, 1))

x_test = x_test.as_matrix().reshape((len(x_test), 1, x_test.shape[1]))
y_test = y_test.as_matrix().reshape((len(y_test), 1, 1))

print("X_train:",x_train.shape)
print("Y_train:",y_train.shape)
print("X_val:",x_test.shape)
print("Y_val:",y_test.shape)


```

```python
epoch_times = 1000

model = Sequential()  
# model.add(Embedding(20, 500, input_length=660))
model.add(GRU(units=100,input_length=x_train.shape[1], input_dim=x_train.shape[2], return_sequences=True,))
model.add(GRU(64, return_sequences=True, dropout=0.2, recurrent_dropout=0.2))
model.add(GRU(32,return_sequences=True,  dropout=0.2, recurrent_dropout=0.2))
model.add(Flatten())
# model.add(Bidirectional(LSTM(units=20,activation='sigmoid',return_sequences=True)))
# model.add(Bidirectional(LSTM(units=20,activation='sigmoid',return_sequences=True)))
model.add(Dense(5,activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
model.fit(x_train, y_train, epochs=epoch_times,validation_data=(x_test, y_test))
```

```python
predict1 = model.predict(x_test)

plt.figure(figsize=(16,9))
plt.title(u'LTSM Model Compare')
plt.plot(range(len(predict1)),predict1[:,:,0],color='navy',linewidth=2,label='LTSM line')
# plt.plot(range(len(predict2)),predict2[:,:,0],color='r',linewidth=2,label='Predict2 line')
# plt.plot(range(len(predict3)),predict3[:,:,0],color='green',linewidth=2,label='Predict3 line')
plt.plot(range(len(y_test)),y_test[:,:,0],color='r',linewidth=1,label='Vaild line')

plt.legend()
fig = plt.gcf()
plt.show()
```

```python
import torch 
import torch.nn as nn 
import numpy as np 
import pandas as pd 
from sklearn import preprocessing 
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt 
import os

# df = pd.read_csv('../../applications/wm5_reshape_less/data/rnn_input_item_seq_with_cluster_integer.csv')
df = pd.read_csv('../../raw-data/wm5-raw.csv')

wm5_grade = pd.read_csv('../../raw-data/wm5-grade.csv')
wm5_id = pd.read_csv('../../raw-data/id.csv')

df = pd.merge( df, wm5_id, on='id', how='right')

df =df.join(wm5_grade.set_index('id'), on='id')

train, test = train_test_split(df, test_size=0.2, shuffle=True)

x_train_raw = train.copy()
x_test_raw = test.copy()

target = 'original'

x_train  = train.drop(['id',target,'Group','Group Project','Exam1','Exam2','Final','Attendance','TA','semester'], axis=1)
x_test  = test.drop(['id',target,'Group','Group Project','Exam1','Exam2','Final','Attendance','TA','semester'], axis=1)

y_train =train[target]
y_test =test[target]


scaler = preprocessing.Normalizer() 
x_train=scaler.fit_transform(x_train) 
x_test=scaler.fit_transform(x_test) 
```

```python
x_train.shape[1]
```

```python
xtrain = torch.Tensor(x_train) 
ytrain = torch.Tensor(y_train)

print(xtrain)
print(ytrain)
```

```python
model = nn.Sequential(
    nn.Linear(x_train.shape[1], 10), 
    nn.ReLU(), 
    nn.Linear(10, 1), 
    nn.ReLU())
```

```python
criterion = torch.nn.L1Loss() 
optimizer = torch.optim.SGD(model.parameters(), lr=0.01) 
for epoch in range(10000): 
    y_pred = model(xtrain) 
    loss = criterion(y_pred, ytrain) 
    print('epoch: ', epoch,' loss: ', loss.item()) 
    optimizer.zero_grad() 
    loss.backward() 
    optimizer.step()
```
