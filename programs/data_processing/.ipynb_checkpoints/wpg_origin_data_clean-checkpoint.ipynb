```python
import pandas as pd
import numpy as np
prefix = 'wpg-5'
origin_data = pd.read_csv('../../raw-data/WPG_data_original_clean.csv',encoding='big5',index_col=0)
# origin_data['Actual AWU']
origin_data.columns
```

```python
df = pd.read_csv('../../applications/%s/data/%s_with_clustered_label_integer.csv' % (prefix,prefix))
df_actual_value = df
df_actual_value['Actual_AWU'] = origin_data['Actual AWU']
df_actual_value['Avail'] = origin_data['Avail.']
df_actual_value['BL<=9WKs'] = origin_data['BL <= 9WKs']
df_actual_value['Backlog'] = origin_data['Backlog']
df_actual_value['DCOH'] = origin_data['DC OH']
df_actual_value['Hub_OH'] = origin_data['Hub OH']
df_actual_value['TTL_OH'] = origin_data['TTL OH']
df_actual_value['FCST_M'] = origin_data['FCST M']

# Backlog_Avail_ratio, AWU_Avial_ratio compute
df_actual_value['AWU_Avial_ratio'] = origin_data['Actual AWU']/origin_data['Avail.']
df_actual_value['Backlog_Avail_ratio'] = origin_data['Backlog']/origin_data['Avail.']
df_actual_value
```

```python
df_actual_value = df_actual_value.dropna()
df_actual_value = df_actual_value.replace(np.inf, 0)
df_actual_value = df_actual_value.replace(-1, 0)
df_actual_value
```

```python
df_actual_value.to_csv('../../applications/%s/data/%s_with_clustered_label_integer_actual_value.csv' % (prefix,prefix),index=False,encoding='utf-8')
```
